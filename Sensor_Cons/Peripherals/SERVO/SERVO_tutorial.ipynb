## Section 1: Servo Motors

### What is a Servo Motor?

A servo motor is a **rotary actuator** that allows for precise control of angular position. Unlike regular DC motors that spin continuously, servos can be positioned to specific angles.

### How Servo Motors Work

**Mechanical Components:**
- **DC Motor**: Provides the rotational force
- **Gearbox**: Reduces speed and increases torque (typically 1:50 to 1:300 ratio)
- **Potentiometer**: Measures the current position
- **Control Circuit**: Compares desired vs actual position and adjusts motor direction

**Control Method:**
- Servos use **Pulse Width Modulation (PWM)** signals
- **50Hz frequency** (20ms period)
- **Pulse width** determines position:
  - 500μs = 0°
  - 1500μs = 90° (center)
  - 2500μs = 180°

### Pin Configuration for STM32F429I-DISC1

**Servo Pinout:**
- **Signal (Orange/Yellow)**: PWM control signal
- **VCC (Red)**: 5V-6V power
- **GND (Brown/Black)**: Ground

**STM32F429I-DISC1 Connections:**
```
Servo Signal Pin → STM32 PWM Pin (e.g., PA8 for TIM1_CH1)
Servo VCC       → 5V power supply
Servo GND       → GND
```

**Important Notes:**
- Servos need their own power supply (don't power from STM32 3.3V)
- Connect grounds together
- Use PWM-capable pins (TIM1, TIM2, TIM3, etc.)

### Basic Servo Code Example

```c
#include "servo.h"

// Initialize servo
SERVO_Handle_t hservo;
TIM_HandleTypeDef htim1;

SERVO_Init(&hservo, &htim1, TIM_CHANNEL_1, GPIOA, GPIO_PIN_8);

// Move to different positions
SERVO_SetAngle(&hservo, 0);    // 0 degrees
HAL_Delay(1000);
SERVO_SetAngle(&hservo, 90);   // 90 degrees (center)
HAL_Delay(1000);
SERVO_SetAngle(&hservo, 180);  // 180 degrees
```

## Section 2: Hardware Setup Tips

### Power Considerations
- **Servos**: Use separate 5V-6V supply, 1A+ per servo
- **STM32**: Logic signals only, don't power servos from board

### Common Issues
- **Servo jitter**: Check power supply stability
- **No movement**: Verify PWM pin and timer configuration
- **Wrong angles**: Calibrate servo range

### Safety
- Never connect servo power to STM32 pins
- Use appropriate fuses
- Start with low angles when testing

## Section 3: Advanced Topics

### Servo Calibration
```c
// Calibrate servo range
SERVO_CalibrateMin(&hservo, 0);    // Set current position as 0°
SERVO_CalibrateMax(&hservo, 180);  // Set current position as 180°
```

### Continuous Movement
```c
// Sweep servo across range
SERVO_Sweep(&hservo, 50);  // Sweep with 50ms delay between steps
```

### PWM Details
The servo driver uses Timer 1 with:
- **Prescaler**: 85 (86MHz → 1MHz timer clock)
- **Period**: 19999 (20000 ticks = 20ms = 50Hz)
- **Pulse**: Variable (500-2500μs for 0°-180°)
